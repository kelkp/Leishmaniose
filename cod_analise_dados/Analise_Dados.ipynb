```python
query = 'select * from View_CasosPorAno'
df = pd.read_sql(query, conn)
QT_Ano=df.to_numpy()
```

##Quantidade de Casos Por Ano

```python
# Barras na vertical
plt.figure(figsize=(10,8))
plt.bar(QT_Ano[:,0],QT_Ano[:,1], color = 'blue')
plt.ylabel("Quantidade de Casos")
plt.xlabel("Ano")
plt.title("Quantidade de Casos vs. Ano")
plt.show()
```
